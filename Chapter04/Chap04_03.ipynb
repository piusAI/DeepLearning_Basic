```python
## Numericial Differential

import numpy as np
import matplotlib.pyplot as plt

# Function : f, x code : x
def numericial_diff(f,x):
  h = 1e-4
  return (f(x+h) - f(x-h) ) / (2 * h)

def function_1(x):
  return 0.01*x**2 + 0.1 * x

def tangent_line(f,x):
  d = numericial_diff(f, x)
  # print(d)
  C = f(x) - d * x
  return lambda t :d*t + C

x=np.arange(-10,10,0.1)

plt.xlabel = ('x')
y=function_1(x)
tf = tangent_line(function_1,2)
y2=tf(x)

plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```
